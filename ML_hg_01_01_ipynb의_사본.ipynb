## 첫 번째 머신러닝 프로그램
---

image.png

```python
length = bream_length+smelt_length
weight = bream_weight+smelt_weight
```

**2차원으로 데이터 생성**

```python
fish_data = [[l, w] for l, w in zip(length, weight)]

print(fish_data)
```

### 정답 준비
---

```python
fish_target = [1]*35 + [0]*14
print(fish_target)
```

### 알고리즘 선택
---
- 사이킷런 설치 : pip install sklearn
- 사이킷런의 KNeighborsClassifier 클래스 사용
- fit() : 학습 시키기
- soore() : 평가 하기
- predict : 예측 하기


```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier()
```

```python
kn.fit(fish_data, fish_target)
```

```python
kn.score(fish_data, fish_target)
```

### k-최근접 이웃 알고리즘
---
- 주위의 데이터로 현재의 데이터 판단
- k는 주위의데이터 개수
- 알고리즘이 모든 데이터를 가지고 있어 데이터가 아주 많은 경우 사용이 어려움

### 새로운 값 예측
---
- 길이=30, 무게=600인 생선 예측
- 빙어 : 0, 도미 : 1

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter(30, 600, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
kn.predict([[30, 600]])
```

### _fit_X 속성
---
- 학습한 fish_data 값

```python
print(kn._fit_X)
```

### _y 속성
---
- target 값

```python
print(kn._y)
```

### n_neighbors 속성 : k값 설정
---
- n_neighbors=49
- k값을 49로 적용 
- 가장 가까운 데이터 49로 사용 하므로 도미가 35개로 빙어 보다 많기 때문에 항상 도미로 예측 하게 됨

```python
kn49 = KNeighborsClassifier(n_neighbors=49)
```

```python
kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
```

```python
print(35/49)
```

### 확인 문제
---
- k 값을 변화 시키면서 확인 해보기

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50):
    # 최근접 이웃 개수 설정
    kn.n_neighbors = n
    # 점수 계산
    score = kn.score(fish_data, fish_target)
    # 100% 정확도에 미치지 못하는 이웃 개수 출력
    if score < 1:
        print(n, score)
        break
```

### 키워드로 끝내는 핵심 포인트
---
- 특성 : 데이터를 표현하는 하나의 성질
- 생선 데이터 특성 : 길이, 무게
- 훈련 : 머신러닝 알고리즘이 데이터에서 규칙을 찾는 과정
- fit() : 사이킷런에서 훈련 메서드
- 모델 : 머신러닝 프로그램에서 알고리즘 자체, 또는 알고리즘이 구현된 객체
- 정확도 :  맞힌 개수 / 전체 데이터 개수

### 핵심 패키지 함수
---
#### matplotlib 
- scatter() : 산점도 그리는 함수

#### scikit-learn
- KNeighborsClassifie() : 분류 모델 클래스
- n_neighbors : 이웃의 개수 지정, 기본값 5
- p : 거리를 재는 방법 지정
- p=1 : 맨해튼 거리 https://bit.ly/man_distance
- p=2 : 유클리디안 거리 https://bit.ly/euc_distance
- n_jobs : CPU 코어수 지정, 기본값 1
